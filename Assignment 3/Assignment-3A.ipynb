````python
## Code goes here
import csv
Firstnames= []
Lastnames=[]
Dirty_Birthmonths=[]
Birthdays=[]
Dirty_Birthyears=[]
Dirty_Courses=[]
Dirty_Phonenumbers=[]
with open('recweekdata.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            Firstnames.append(row[0])
            Lastnames.append(row[1])
            Dirty_Birthmonths.append(row[2])
            Birthdays.append(row[3])
            Dirty_Birthyears.append(row[4])
            Dirty_Courses.append(row[5])
            Dirty_Phonenumbers.append(row[6])
            line_count += 1
            
    print(f'Processed {line_count} lines.')
    print(Birthmonths)
````

````python
#Birthmonth cleaning
#used technique in Day 14 discussions
Unique_Dirty_Birthmonths=set(Dirty_Birthmonths)
Birthmonth_Dict = {name:None for name in Unique_Dirty_Birthmonths}

for i in Birthmonth_Dict:
    cleaned_name = input(i+": "+"Enter name: ")
    Birthmonth_Dict[i]=cleaned_name
Birthmonth_Dict
````

````python
import re
Cleaned_Birthmonths = [re.sub(r''.join(i),r''.join(Birthmonth_Dict[i]),i) for i in Dirty_Birthmonths]
Cleaned_Birthmonths
````

````python
#Cleaning birthyears
Cleaned_Birthyears=[]
#all you have to do is add the first 2 digits or remove them to clean so I'll just settle for adding because I think it looks better and avoids confusion
for i in Dirty_Birthyears:
    if len(i)==4:
        Cleaned_Birthyears.append(i)
    else:
        if i[0] == 0:
            Cleaned_Birthyears.append('20'+i)
        else:
            Cleaned_Birthyears.append('19'+i)
Cleaned_Birthyears
````

````python
#Cleaning Courses
#also sir you just said to clean the data which I will but I kind of want to shorten the names instead of making it the course title
#Also same technique done in Day 14
Unique_Dirty_Courses=set(Dirty_Courses)
Courses_Dict = {name:None for name in Unique_Dirty_Courses}

for i in Courses_Dict:
    cleaned_name = input(i+": "+"Enter name: ")
    Courses_Dict[i]=cleaned_name
Cleaned_Courses = [re.sub(r''.join(i),r''.join(Courses_Dict[i]),i) for i in Dirty_Courses]
Cleaned_Courses
````

````python
#Cleaning phonenumbers
#Similar method to years except using regex to find the 9
Cleaned_Phonenumbers=[]
for i in Dirty_Phonenumbers:
    Cleaned_Phonenumber = re.search(r'9.+', i).group()
    Cleaned_Phonenumbers.append("0"+Cleaned_Phonenumber)
Cleaned_Phonenumbers
````

````python
#Have to combine all the cleaned and old used data together
Cleaned_Data=zip(Firstnames,Lastnames,Cleaned_Birthmonths,Birthdays,Cleaned_Birthyears,Cleaned_Courses,Cleaned_Phonenumbers)
Cleaned_Data=list(Cleaned_Data)
Cleaned_Data
columns = ["Firstnames","Lastnames","Cleaned_Birthmonths","Birthdays","Cleaned_Birthyears","Cleaned_Courses","Cleaned_Phonenumbers"]
with open('cleanrecweekdata.csv', 'w') as csvFile:
    writer = csv.writer(csvFile)
    writer.writerow(columns)
    for i in Cleaned_Data:
            writer.writerow(i)
````
