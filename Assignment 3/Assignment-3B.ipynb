````python
## Code goes here
import pandas as pd
import csv
import re

````

````python
#Making lists that turn into the columns
Firstnames= []
Lastnames=[]
Dirty_Birthmonths=[]
Birthdays=[]
Dirty_Birthyears=[]
Dirty_Courses=[]
Dirty_Phonenumbers=[]

with open('recweekdata.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            Firstnames.append(row[0])
            Lastnames.append(row[1])
            Dirty_Birthmonths.append(row[2])
            Birthdays.append(row[3])
            Dirty_Birthyears.append(row[4])
            Dirty_Courses.append(row[5])
            Dirty_Phonenumbers.append(row[6])
            line_count += 1
            
    print(f'Processed {line_count} lines.')

````

````python
#turning the genders into a list so that I don't need to join two dataframes 
Genders=[]             
with open('studentprofiles.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            Genders.append(row[2])
            line_count += 1
````

````python
df = pd.DataFrame({"firstname":Firstnames, "lastname":Lastnames, "birthmonth":Dirty_Birthmonths, 
                   "birthday": Birthdays, "birthyear": Dirty_Birthyears, "course":Dirty_Courses,"phonenumber":Dirty_Phonenumbers,
                  "gender":Genders})
df
````

````python
#Brute forcing replacements
df.loc[:,"phonenumber"].replace(r"^\+63","",regex=True, inplace=True)
df.loc[:,"phonenumber"].replace(r"^ ","",regex=True, inplace=True)
carrier = {"917":"Globe","918":"Smart", "927":"Globe"}

df["carrier"] = df.loc[:,"phonenumber"].str[0:3].apply(lambda x: carrier[x])
df.loc[:,"birthyear"].replace(r"^97","1997",regex=True, inplace = True)
df.loc[:,"birthyear"].replace(r"^98","1998",regex=True, inplace = True)
df.loc[:,"birthyear"].replace(r"^99","1999",regex=True, inplace = True)

df.loc[:,"course"].replace(r"^Management Engineering","BS ME",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^Management","BS MGT",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^ME","BS ME",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^ITE","BS ITE",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^CTM","BS CTM",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^LM","BS LM",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^MGT","BS MGT",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^Information Technology Entrepreneurship","BS ITE",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^Communications Technology","BS CTM",regex=True, inplace = True)
df.loc[:,"course"].replace(r"^Legal Management","BS LM",regex=True, inplace = True)

df.loc[:,"birthmonth"].replace(r"^10","October",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^11","November",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^12","December",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^1","January",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^2","February",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^3","March",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^4","April",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^5","May",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^6","June",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^7","July",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^8","August",regex=True, inplace = True)
df.loc[:,"birthmonth"].replace(r"^9","September",regex=True, inplace = True)

df.loc[:,"birthmonth"].replace(r"^Julyy","July",regex=True, inplace = True)
df["birthmonth"]= df.loc[:,"birthmonth"].str[0:3]
````

````python
df.to_csv("cleanrecweekdata3b.csv")
````
